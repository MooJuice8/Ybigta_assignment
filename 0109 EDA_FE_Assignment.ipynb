### 5) Feature Engineering

feature에 대한 전처리를 하기 이전에 결측치가 있는 column을 모두 제거했는데, 다시 초기 상태로 되돌리고 진행하겠습니다.

```python
df = pd.read_csv('Data/airbnb_listings.csv', index_col=0)
```

- #### <font color='orange'> Q12. 결측치에 대한 처리를 하고 싶습니다.
#### <font color='orange'> 주어진 모든 column들에 대하여 결측치가 25%이상인 경우에는 column을 drop하고 10%이상 25%미만인 경우에는 결측치가 있는 row만 drop하고, 10%미만인 경우에는 중앙값으로 대체해주세요.

```python
#Q12
# Analyze the percentage of missing values for each column
missing_values_percent = df.isnull().mean() * 100

# Columns to drop (more than 25% missing values)
columns_to_drop = missing_values_percent[missing_values_percent > 25].index

# Columns for which rows will be dropped (between 10% and 25% missing values)
columns_rows_drop = missing_values_percent[(missing_values_percent >= 10) & (missing_values_percent <= 25)].index

# Drop columns with more than 25% missing values
df_dropped_columns = df.drop(columns=columns_to_drop)

# Drop rows with missing values in certain columns (10% to 25% missing values)
df_dropped_rows = df_dropped_columns.dropna(subset=columns_rows_drop)

# Identify numeric columns in the dataframe
numeric_columns = df_dropped_rows.select_dtypes(include=['number']).columns

# Replace missing values only in numeric columns with their median
df_final_numeric_median = df_dropped_rows.fillna(df_dropped_rows[numeric_columns].median())

# Check the result after this correction
df_final_numeric_median.info()
```

- #### <font color='orange'> Q13. 앞서 진행한 EDA를 바탕으로 FE를 진행하고자 합니다..
#### <font color='orange'> 총 3개의 FE를 진행하시고, 해당 FE를 진행한 이유에 대해 **간단히** 작성해주세요.


```python
#Q13-1)
# 1. Creating 'Price per Accommodate' Feature
# gives a sense of the cost efficiency of the listing, providing a normalized measure of price that accounts for the number of guests it can accommodate. 
# This is useful in comparing listings and understanding pricing strategies.
df_final_numeric_median['price_per_accommodate'] = df_final_numeric_median['price'] / df_final_numeric_median['accommodates']
df_final_numeric_median['price_per_accommodate']
```

```python
#Q13-2)
# 2. Review Scores Ratio
# Assuming the dataset has review scores like 'review_scores_rating', 'review_scores_accuracy', etc.
review_score_columns = ['review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness', 'review_scores_checkin', 'review_scores_communication', 'review_scores_location', 'review_scores_value']
df_final_numeric_median['review_scores_combined'] = df_final_numeric_median[review_score_columns].mean(axis=1)
df_final_numeric_median['review_scores_combined']
```

```python
#Q13-3)
# 3. Boolean Conversion of 'Instant Bookable'
# The 'instant_bookable' column has been converted into a boolean format (1 for 'true' and 0 for 'false')
df_final_numeric_median['instant_bookable_bool'] = df_final_numeric_median['instant_bookable'].map({'t': 1, 'f': 0})
df_final_numeric_median['instant_bookable_bool']
```
